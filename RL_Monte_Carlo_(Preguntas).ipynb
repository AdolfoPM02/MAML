### Ejercicio 6: Métodos de Monte Carlo "From Scratch"

**Contexto:**
Avanzamos hacia problemas con estado completo de RL usando Procesos de Decisión de Markov (MDP). Trabajaremos con el entorno `Blackjack-v1` de `Gymnasium`.

**Tu Tarea:**

1. Implementa el algoritmo de **Predicción de Monte Carlo Primera-Visita (First-Visit MC Prediction)** para evaluar la política de "plantarse a los 20 o 21, y pedir carta en caso contrario".
2. Implementa **Control de Monte Carlo con $\epsilon$-soft policies** (First-Visit MC Control) para encontrar la política óptima.
3. Dibuja el mapa de calor (Heatmap) de la función de valor óptima $V^*(s)$ descubierta y la política óptima $\pi^*(s)$.
