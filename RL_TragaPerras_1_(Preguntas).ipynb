### Ejercicio 1.bis: El Problema No Estacionario ($\epsilon$-Greedy con Step-Size Constante)

**Contexto:**
En el mundo real, la verdadera recompensa de una acción raramente es estática. Imagina un sistema de recomendación donde los gustos del usuario cambian. Para simular esto, el valor real de los brazos $q(a)$ no será constante.
El tiempo que dura una configuración de $q(a)$ (una "época") seguirá una distribución exponencial. Cuando una época termina, todos los valores $q(a)$ se vuelven a sortear desde la distribución $\mathcal{N}(0, 1)$.

El algoritmo $\epsilon$-greedy clásico usa un promedio muestral: $Q_{n+1} = Q_n + \frac{1}{n}(R_n - Q_n)$. A medida que $n$ crece, las nuevas recompensas apenas modifican la estimación, lo cual es fatal en entornos no estacionarios. La solución es usar un *step-size* constante $\alpha \in (0, 1]$:
$Q_{n+1} = Q_n + \alpha(R_n - Q_n)$

**Tu Tarea:**

1. Crea un nuevo entorno `NonStationaryBanditEnvironment` donde la duración de cada "época" se extraiga de una variable aleatoria exponencial con media de 1000 pasos (es decir, la tasa $\lambda = 1/1000$). En cada cambio de época, sortea nuevos $q(a)$.
2. Implementa un agente $\epsilon$-greedy que utilice un ratio de aprendizaje (step-size) constante $\alpha$.
3. Simula el entorno y compara el $\epsilon$-greedy clásico (promedio muestral) frente al $\epsilon$-greedy de step-size constante ($\alpha = 0.1$).

### Ejercicio 3: Gradient Bandits

**Contexto:**
En lugar de estimar el valor de la acción, el algoritmo Gradient Bandit aprende una preferencia numérica $H_t(a)$ para cada acción. Las probabilidades de selección se basan en una distribución softmax.
Regla de actualización:
$H_{t+1}(A_t) = H_t(A_t) + \alpha (R_t - \bar{R}_t)(1 - \pi_t(A_t))$
Para todas las demás acciones $a \neq A_t$:
$H_{t+1}(a) = H_t(a) - \alpha (R_t - \bar{R}_t)\pi_t(a)$

**Tu Tarea:**

### Ejercicio 5: Thompson Sampling

**Contexto Teórico:**
Hasta ahora hemos visto métodos basados en estimaciones puntuales (promedios). **Thompson Sampling** es un enfoque *Bayesiano* que no hemos visto en la parte teorica. En lugar de guardar un solo número para la recompensa esperada de una acción, guardamos una **distribución de probabilidad** completa que representa nuestra incertidumbre sobre el verdadero valor de esa acción.

Para entenderlo fácilmente, cambiaremos temporalmente a un **Bandido de Bernoulli**, donde la recompensa es binaria (1 éxito, 0 fracaso).

1. **Prior:** Inicialmente, no sabemos nada sobre la probabilidad de éxito de un brazo. Representamos esto con una distribución uniforme, que en estadística Bayesiana equivale a una distribución $Beta(\alpha=1, \beta=1)$.
2. **Sampling:** En cada paso $t$, en lugar de elegir el valor medio o calcular un intervalo de confianza matemático, *muestreamos* un valor aleatorio de la distribución $Beta$ de cada brazo. Elegimos el brazo con la muestra más alta.
3. **Actualización (Posterior):** Si el brazo nos da una recompensa de 1 (éxito), sumamos 1 al parámetro $\alpha$ de ese brazo. Si nos da 0 (fracaso), sumamos 1 al parámetro $\beta$.

A medida que jugamos, las distribuciones Beta se vuelven más estrechas (menos incertidumbre) alrededor de la verdadera tasa de éxito.

**Tu Tarea:**

1. Crea un entorno `BernoulliBandit` donde cada brazo tiene una probabilidad real de éxito $p(a) \sim U(0, 1)$.
2. Implementa el agente `ThompsonSamplingAgent` manteniendo matrices para los parámetros $\alpha$ y $\beta$ de cada brazo.
3. Simula 2000 pasos y compara visualmente el rendimiento de Thompson Sampling frente a un agente $\epsilon$-greedy convencional adaptado a este entorno binario.
